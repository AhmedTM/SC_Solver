```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data1 = pd.read_csv("./datasets/3_dataset_1.csv")
data2 = pd.read_csv("./datasets/3_dataset_2.csv")

data1_double = pd.read_csv("./output/data1_double_pointsNP.csv")
data1_4times = pd.read_csv("./output/data1_4times_pointsNP.csv")

data2_double = pd.read_csv("./output/data2_double_pointsNP.csv")
data2_4times = pd.read_csv("./output/data2_4times_pointsNP.csv")
```
